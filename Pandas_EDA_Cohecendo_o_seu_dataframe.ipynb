# Pandas EDA - Conhecendo o seu dataframe

Neste notebook, o objetivo fazer uma análise exploratória básica utilizando alguns atributos e métodos do Pandas. Aqui, não utilizaremos gráficos e coisas do tipo, assunto para um próximo notebook.

O dataset utilizado no exemplo foi obtido em https://www.kaggle.com/samawel97/supermarket-sales-analysis-visualisation

Vamos transformar o arquivo csv em um objeto do Pandas e, para confirmar que tudo deu certo, já vamos pedir para que o Pandas nos mostre as primeiras linhas do dataset.

```
sales = pd.read_csv("supermarket_sales - Sheet1.csv") #se estiveres trabalhando em máquina local, sera necessário colocar o caminho completo do arquivo
sales.head()
```

Podemos informar um parâmetro para 'head()' nos mostrar um número específico de linhas que queremos ver no nosso dataset

```
sales.head(15)
```

Para ver os últimos registros de um dataset, use tail()

```
sales.tail(10)
```

Use info() para ver algumas informações importantes do seu dataframe

```
sales.info()
```

columns lista todas as colunas do dataframe; values() devolve um array de listas com os registros

```
sales.columns
```

```
sales.values
```

Com describe(), consigo ver alguns dados estatísticos gerais, como desvio padrão média, quartis e valores de máximo e mínimo.

```
sales.describe()
```

Para localizar um registro a partir de um índice, basta usar iloc[índice]

```
a = sales.iloc[22]
a
```

Slicing em Pandas segue o mesmo padrão do Python. dataframe[ínicio:fim]. Para o exemplo abaixo, queremos exibir todos os registros entre 100 e 125

```
sales[100:125]
```

Podemos utilizar as funções isnull() e notnull() para verificar a (in)existencia de valores nulos no dataframe.

```
pd.isnull(sales)
```

```
pd.notnull(sales)
```

```
Também é possível fazer verificações rápidas através de operações lógicas. Abaixo, verificamos para quais índices o valor Quantity é maior que 3
```

```
sales.Quantity > 3
```

Para finalizar, também é possível selecionar colunas específicas e visualizar dados.

```
print(sales[['Branch','Gender','Payment']])
```
